```python
import numpy as np
import math
def grad_desc(x,y):
    m_curr = b_curr=0
    iterations = 1000
    n=len(x)
    learning_rate = 0.0001
    index=0
    for i in range(iterations):
        y_predicted = m_curr*x +b_curr
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted)) #derivative of loss function with respect to m
        bd = -(2/n)*sum(y-y_predicted)     #derivative of loss function with respect to b
        m_curr = m_curr-learning_rate*md
        b_curr = b_curr-learning_rate*bd
        
        print(f"m{m_curr}, b{b_curr},cost{cost}, iterations{i}")
        if math.isclose(cost,(1/n)*sum([val**2 for val in (y-(m_curr*x+b_curr))])):
            print("Converged")
            index=i
            break
    print(i)

x=np.array([92,56,88,70,80,49,65,35,66,67])
y=np.array([98,68,81,80,83,52,66,30,68,73])
grad_desc(x,y)
```
