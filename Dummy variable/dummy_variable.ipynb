# One Hot encoding


```python
import pandas as pd
df =pd.read_csv("dataset/homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged_dataset = pd.concat([df,dummies],axis='columns')
merged_dataset
```

```python
final = merged_dataset.drop(['town','west windsor'],axis ='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x=final.drop(['price'],axis=1)
x

```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,1,0],[30000,0,1]])
```

```python
model.score(x,y)
```
